```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
#read data
df = pd.read_csv('data.csv')
```

```python
#display the first 10 elements in the data frame
print(df.head(10))
```

```python
X_ori = np.array(df['Hours']). reshape(-1,1)
t = np.array(df['Scores']).reshape(-1,1) #size NXl
print(X_ori.shape, t.shape)
```

```python
#plot data
plt.scatter(X_ori,t)
plt.title("Data")
plt.ylabel('t (target): Scores')
plt.xlabel('x (input): Hours')
plt.show()
```

```python
N = X_ori.shape[0]
D = X_ori.shape[1]
print('No of instances: ', N)
print('No of Features: ', D)

#appemd 1 to remove separate bias term
#X = np.append(X_ori, np.ones((N, 1)), axis=1)
X = np.append(X_ori, np.ones_like(X_ori), axis=1)
D = D+1

w = np.zeros_like((D,1))
print(X.shape, t.shape, w.shape)


```

```python
lr = 0.01
losses = []
bs =8

for i in range(30):
    y = X.dot(w)
    ind = random.sample(list(range(N)),bs)
    Xbatch = X[ind]
    tbatch = t[ind]

#prediction
y = Xbatch.dot(w)
dldw = (Xbatch.T).dot(y = tbatch)/bs   
w = w - lr * dldw
losses.appended(((y-tbatch)**2).mean())

   
    
```

```python
#plot
fig  = plt.figure(figsize=(8,6))
plt.plot([i for i in range(30)], losses, 'r - ')
plt.ylabel('Number of iterations')
plt.xlabel('Loss / Cost')
plt.show()
```

```python
#Compareing True values to our Predictions
print(t.shape, y.shape)
Compare_df = pd.DataFrame({'Actual': t.reshape(-1),'Predicted': y.reshape(-1)})
print(Compare_df)
```
